```python
#PROJECT 2- LEVEL-1
### Customer Segmentation Analysis
import pandas as pd
df=pd.read_csv("C:/Users/shiva/Documents/oasis/ifood_df.csv")
df
```

```python
print("First few rows of the dataset:")
print(df.head())


print("\nSummary of the dataset:")
print(df.info())


print("\nStatistical summary of the dataset:")
print(df.describe())

# Step 3: Clean the dataset

print("\nMissing values in each column:")
print(df.isnull().sum())


df['Income'] = df['Income'].fillna(df['Income'].mean())

```

```python
print("\nMissing values after cleaning:")
print(df.isnull().sum())
```

```python
# Step 4: Calculate Descriptive Statistics
# Calculate average purchase value
average_purchase_value = df['MntTotal'].mean()
print(f"\nAverage purchase value: {average_purchase_value:.2f}")

# Calculate frequency of purchases per customer (based on number of accepted campaigns)
purchase_frequency = df['AcceptedCmpOverall']
print("\nPurchase frequency per customer (number of accepted campaigns):")
print(purchase_frequency.value_counts())

```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Select relevant features
features = ['Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts',
            'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'MntTotal']

# Normalize the features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Step 3: Apply K-means clustering
# Determine the optimal number of clusters using the elbow method
inertia = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow method
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()

# Apply K-means with the optimal number of clusters (let's assume k=4 based on the elbow plot)
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df_scaled)

# Step 4: Visualize the customer segments
# Plot clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Income', y='MntTotal', hue='Cluster', palette='viridis')
plt.title('Customer Segments by Income and Total Spend')
plt.xlabel('Income')
plt.ylabel('Total Spend')
plt.legend(title='Cluster')
plt.show()


# Step 5: Analyze characteristics of each segment and provide insights
# Group by cluster and calculate mean values
cluster_summary = df.groupby('Cluster')[features].mean()
print("\nCluster Summary:")
print(cluster_summary)

# Add more analysis as needed
# Example: Distribution of 'Kidhome' and 'Teenhome' in each cluster
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cluster', y='Kidhome', palette='viridis')
plt.title('Distribution of Kidhome in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Kidhome')
plt.show()

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='Cluster', y='Teenhome', palette='viridis')
plt.title('Distribution of Teenhome in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Teenhome')
plt.show()

```

```python
# Step 6: Insights and Recommendations
for cluster in range(4):
    print(f"\nCluster {cluster} characteristics:")
    print(cluster_summary.loc[cluster])
    print(f"Number of customers: {len(df[df['Cluster'] == cluster])}")

    # Example insights
    if cluster_summary.loc[cluster]['Income'] > df['Income'].mean():
        print("This cluster has a higher than average income.")
    else:
        print("This cluster has a lower than average income.")
    
    if cluster_summary.loc[cluster]['MntTotal'] > df['MntTotal'].mean():
        print("This cluster spends more than the average.")
    else:
        print("This cluster spends less than the average.")
    
```
